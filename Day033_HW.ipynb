```python
import requests
from bs4 import BeautifulSoup
```

```python
#1.需增加header的user agent
#2.需加入cookies，並比較登入前後cookies有缺的部分，補到request內送出請求即可

headers = {
    'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.117 Safari/537.36'
}
cookies = {
    'dotcom_user': 'cafemato',
    'logged_in': 'yes',
    'user_session': 'J2vUWj8v92fxPw_SeQbIHOpo4q39gdC8cZ0XniPNwx7e5ZcZ'
}
```

```python
rs = requests.session()
res = rs.get('https://github.com/new', headers=headers, cookies=cookies)
soup = BeautifulSoup(res.text,'html.parser')
print(soup.text)
```
